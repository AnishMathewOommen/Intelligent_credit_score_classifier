```python
# Import necessary libraries
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Read the dataset
data = pd.read_csv('dataset/credit_score.csv')
data.head()
```

```python
# Check the column names
data.columns
```

```python
# Remove the unwanted columns
data = data.drop(columns=['ID','Customer_ID','Month','Name','Type_of_Loan','SSN','Occupation','Credit_History_Age'])
data.head()
```

Data Cleaning


```python
# Convert columns to appropriate data types and clean values
data['Age'] = data['Age'].astype(str).str.replace('_', '').astype(float).round().astype(int)
data['Annual_Income'] = data['Annual_Income'].astype(str).str.replace('_', '').astype(float)
data['Num_of_Loan'] = data['Num_of_Loan'].astype(str).str.replace('_', '').astype(float)
data['Num_of_Delayed_Payment'] = pd.to_numeric(data['Num_of_Delayed_Payment'], errors='coerce')  # Convert to numeric, coercing errors

# Encode categorical values in 'Credit_Score'
data['Credit_Score'] = data['Credit_Score'].replace(["Poor", "Standard", "Good"], [0, 1, 2])

# Clean and convert financial columns to numeric types
data['Monthly_Balance'] = data['Monthly_Balance'].astype(str).str.replace('_', '').astype(float)
data['Monthly_Inhand_Salary'] = data['Monthly_Inhand_Salary'].astype(float)  # Already numeric
data['Amount_invested_monthly'] = data['Amount_invested_monthly'].astype(str).str.replace('_', '').astype(float)
data['Outstanding_Debt'] = data['Outstanding_Debt'].astype(str).str.replace('_', '').astype(float)

# Clean 'Payment_Behaviour' and convert to NaN where applicable
data['Payment_Behaviour'] = data['Payment_Behaviour'].replace('!@9#%8', np.nan)  # Replace invalid entries with NaN
data['Payment_of_Min_Amount'] = data['Payment_of_Min_Amount'].replace({'NM': 0, 'Yes': 1, 'No': 0}).astype(int)  # Encode Yes/No

# Clean 'Credit_Mix' and convert to numerical values
data['Credit_Mix'] = data['Credit_Mix'].replace('_', np.nan)  # Replace placeholders with NaN
data['Credit_Mix'] = data['Credit_Mix'].replace(["Good", "Standard", "Bad"], [2, 1, 0])  # Encode credit mix categories
data['Changed_Credit_Limit'] = data['Changed_Credit_Limit'].astype(str).replace('_', np.nan).astype(float)  # Convert to float

# Display the cleaned data structure for verification
print(data.info())
```

```python
# Check for missing values
data.isnull().sum()
```

```python
# Fill missing values
data.fillna(method='ffill', inplace=True)
data.fillna(method='bfill', inplace=True)
data.isnull().sum()
```

```python
# Remove outliers
Q1 = data['Age'].quantile(0.25)
Q3 = data['Age'].quantile(0.75)
IQR = Q3 - Q1
data = data[~((data['Age'] < (Q1 - 1.5 * IQR)) | (data['Age'] > (Q3 + 1.5 * IQR)))]

sns.boxplot(x=data['Age'])
plt.xlabel('Age')
plt.ylabel('Count')
plt.show() 
```

```python
# Encode categorical values
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data["Payment_Behaviour"] = le.fit_transform(data["Payment_Behaviour"])
data.info()
```

Feature Selection

```python
# Check for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

col_list = []
for col in data.columns:
    if data[col].dtypes != 'object' and col != 'Credit_Score':
        col_list.append(col)
X = data[col_list]

vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]

print(vif_data)

# Selecting the features using VIF. VIF should be less than 5. Here,all features have VIF value less than 5, So we will select all the features.
```

Logistic Regression

```python
# Split the data into training and testing sets
X = data.drop(columns=['Credit_Score'])
Y = data['Credit_Score']

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=30)
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Import the Logistic Regression model from scikit-learn
from sklearn.linear_model import LogisticRegression


# Import accuracy_score from scikit-learn to evaluate the model's performance
from sklearn.metrics import accuracy_score


# Initialize the Logistic Regression model
logistic_regression_obj = LogisticRegression()

# Fit the model to the training data
logistic_regression_obj.fit(x_train, y_train)

# Make predictions on the test data
y_pred = logistic_regression_obj.predict(x_test)

# Calculate and print the accuracy of the model
logistic_regression_accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression Accuracy: ", logistic_regression_accuracy)

# Create a DataFrame to compare actual and predicted values, displaying the first 10 rows
comparison_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(comparison_df.head(10))
```

```python
# Import accuracy_score from scikit-learn to evaluate the model's performance
from sklearn.metrics import accuracy_score

# Calculate and print the accuracy of the model
logistic_regression_accuracy = accuracy_score(y_test, y_pred)
print("Logistic Regression Accuracy: ", logistic_regression_accuracy)

# Create a DataFrame to compare actual and predicted values, displaying the first 10 rows
comparison_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(comparison_df.head(10))
```

```python
# Import the Decision Tree Classifier from scikit-learn
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
decision_tree_obj = DecisionTreeClassifier()

# Fit the model to the training data
decision_tree_obj.fit(x_train, y_train)

# Make predictions on the test data
y_pred = decision_tree_obj.predict(x_test)

# Calculate the accuracy of the model
decision_tree_classifier_accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy in a formatted way
print(f"Decision Tree Classifier Accuracy: {decision_tree_classifier_accuracy:.2f}")

# Create a DataFrame to compare actual and predicted values, displaying the first 10 rows
comparison_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(comparison_df.head(10))
```

```python
# Import the Random Forest Classifier from scikit-learn
from sklearn.ensemble import RandomForestClassifier

# Initialize the Random Forest Classifier
random_forest_obj = RandomForestClassifier()

# Fit the model to the training data
random_forest_obj.fit(x_train, y_train)

# Make predictions on the test data
y_pred = random_forest_obj.predict(x_test)

# Calculate the accuracy of the model
random_forest_classifier_accuracy = accuracy_score(y_test, y_pred)

# Print the accuracy in a formatted way
print(f"Random Forest Classifier Accuracy: {random_forest_classifier_accuracy:.2f}")

# Create a DataFrame to compare actual and predicted values, displaying the first 10 rows
comparison_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(comparison_df.head(10))
```

```python
# Import the SVM classifier from scikit-learn
from sklearn.svm import SVC

# SVM Classification
# Initialize the SVM classifier with a polynomial kernel
svm_obj = SVC(kernel='poly')  

# Fit the model to the training data
svm_obj.fit(x_train, y_train)

# Make predictions on the test data
y_pred_svm = svm_obj.predict(x_test)

# Calculate and print the accuracy of the SVM model
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Classifier Accuracy: {accuracy_svm:.2f}")

# Create a DataFrame to compare actual and predicted values, displaying the first 10 rows
comparison_df_svm = pd.DataFrame({"Actual": y_test, "Predicted": y_pred_svm})
print(comparison_df_svm.head(10))
```

```python
# List of classifiers and their corresponding accuracies
classifiers = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'SVM']
accuracies = [
    logistic_regression_accuracy,  
    decision_tree_classifier_accuracy,         
    random_forest_classifier_accuracy,         
    accuracy_svm                    
]

# Creating a bar plot for accuracy
plt.figure(figsize=(10, 6))
bars = plt.bar(classifiers, accuracies, color=['#4CAF50', '#2196F3', '#FF9800', '#F44336'])

# Adding labels and title
plt.xlabel('Classifiers', fontsize=14)
plt.ylabel('Accuracy', fontsize=14)
plt.title('Classifier Accuracy Comparison', fontsize=16)

# Adding accuracy values on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom', fontsize=12)

# Display the plot
plt.ylim(0, 1)  # Set y-axis limits to 0 to 1 for better visualization
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
